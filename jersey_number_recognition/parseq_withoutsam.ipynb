```python
import cv2
import numpy as np
import pandas as pd
import pickle
from segment_anything import SamPredictor, sam_model_registry
from jersey_number_pipeline import JerseyNumberPipeline

# SAM modelini yükle
sam = sam_model_registry["vit_h"](checkpoint="/content/jersey_number_pipeline/segment_anything/sam_vit_h_4b8939.pth")
predictor = SamPredictor(sam)

def segment_player(image, bbox):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    predictor.set_image(image_rgb)
    masks, _, _ = predictor.predict(box=bbox, multimask_output=False)
    return masks[0]

# Pipeline’ı yükle
pipeline = JerseyNumberPipeline(
    legible_classifier_path='/content/jersey-number-pipeline/models/resnet34_legibility.pth',
    parseq_model_path='/content/jersey-number-pipeline/models/parseq_jersey_soccer.pth',
    reid_model_path='/content/jersey-number-pipeline/reid/centroids-reid/models/centroids_reid.pth',
    pose_model_path='/content/jersey-number-pipeline/pose/ViTPose/checkpoints/vitpose.pth'
)

# Sahne zamanlarını oku, ilk satırı atla
scenes_df = pd.read_csv('/content/drive/MyDrive/match_video-Scenes.csv', header=1)
scenes = scenes_df[['Start Time (seconds)', 'End Time (seconds)']].values

# Videoyu oku
video_path = '/content/drive/MyDrive/videos/match_video.mp4'
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Her sahne için tracklet oluştur
tracklets = []
for idx, (start_time, end_time) in enumerate(scenes):
    print(f"Sahne {idx + 1} işleniyor: {start_time}s - {end_time}s")

    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    frames = []
    for frame_idx in range(start_frame, end_frame):
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    # Sahne içinde tracklet oluştur
    scene_tracklets = pipeline.process_frames(frames)

    # SAM ile segmentasyon yap
    for tracklet in scene_tracklets:
        for frame_idx, result in tracklet.items():
            bbox = result['bbox']
            mask = segment_player(frames[frame_idx - start_frame], bbox)
            result['mask'] = mask

    tracklets.append(scene_tracklets)

    # Her sahneyi işledikten sonra tracklet’leri Drive’a kaydet
    with open(f'/content/drive/MyDrive/tracklets_scene_{idx}.pkl', 'wb') as f:
        pickle.dump(scene_tracklets, f)

cap.release()
print("Tracklet’ler oluşturuldu ve Drive’a kaydedildi.")
```

```python
import cv2
import pandas as pd
import numpy as np
import pickle
import glob
import torch
import torchvision.transforms as transforms
from PIL import Image

# PARSeq ve Legibility Classifier için gerekli modülleri yükle
import sys
sys.path.append('/content/jersey_number_pipeline/parseq')
from parseq.models.parseq import PARSeq
from parseq.utils import load_checkpoint

# Bağımlılıkları yükle (önceden yapılmadıysa)
!pip install torch==2.3.1+cu121 torchvision==0.18.1+cu121 torchaudio==2.3.1+cu121 -f https://download.pytorch.org/whl/torch_stable.html
!pip install pytorch-lightning --upgrade
!pip install einops mlflow opencv-python

# Cihaz ayarı (GPU varsa kullan)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# PARSeq modelini yükle
parseq_model_path = '/content/jersey_number_pipeline/models/parseq_jersey_soccer.pth'
parseq = PARSeq.load_from_checkpoint(parseq_model_path).to(device)
parseq.eval()

# Legibility Classifier modelini yükle (ResNet34 tabanlı)
legibility_model_path = '/content/jersey_number_pipeline/models/resnet34_legibility.pth'
legibility_classifier = torch.load(legibility_model_path, map_location=device)
legibility_classifier.eval()

# Görüntü ön işleme için transform
transform = transforms.Compose([
    transforms.Resize((32, 128)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def process_jersey_region(frame, bbox):
    """Bounding box içindeki forma bölgesini kırp ve ön işle."""
    x, y, w, h = bbox
    jersey_region = frame[y:y+h, x:x+w]
    jersey_region = cv2.cvtColor(jersey_region, cv2.COLOR_BGR2RGB)
    jersey_region = Image.fromarray(jersey_region)
    jersey_region = transform(jersey_region).unsqueeze(0).to(device)
    return jersey_region

def is_legible(jersey_region):
    """Forma numarasının okunabilirliğini kontrol et."""
    with torch.no_grad():
        logits = legibility_classifier(jersey_region)
        prob = torch.softmax(logits, dim=1)
        legible_prob = prob[:, 1].item()  # Okunabilir olma olasılığı
        return legible_prob > 0.5  # Eşik: %50

def read_jersey_number(jersey_region):
    """Forma numarasını PARSeq ile oku."""
    with torch.no_grad():
        logits = parseq(jersey_region)
        pred = parseq.decode(logits)
        return pred[0] if pred else 'N/A'

# Tracklet’ları yükle
tracklet_files = glob.glob('/content/drive/MyDrive/tracklets_scene_*.pkl')
tracklets = []
for file in tracklet_files:
    with open(file, 'rb') as f:
        scene_tracklets = pickle.load(f)
        tracklets.append(scene_tracklets)

print(f"{len(tracklets)} sahne için tracklet’lar yüklendi.")

# Video ve sahne bilgilerini oku
video_path = '/content/drive/MyDrive/videos/match_video.mp4'
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

scenes_df = pd.read_csv('/content/drive/MyDrive/match_video-Scenes.csv')
scenes = scenes_df[['Start Time (seconds)', 'End Time (seconds)']].values

# Çıktı CSV dosyası
output_csv = "/content/jersey_numbers.csv"
with open(output_csv, "w") as f:
    f.write("second,jersey_number,bbox\n")

# Her sahne için tracklet’ları işle
for idx, (start_time, end_time) in enumerate(scenes):
    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    scene_tracklets = tracklets[idx]

    for frame_idx in range(start_frame, end_frame):
        ret, frame = cap.read()
        if not ret:
            break

        second = frame_idx / fps
        for tracklet in scene_tracklets:
            if frame_idx in tracklet:
                result = tracklet[frame_idx]
                bbox = result['bbox']

                # Forma bölgesini kırp ve ön işle
                jersey_region = process_jersey_region(frame, bbox)

                # Okunabilirlik kontrolü
                if is_legible(jersey_region):
                    # Forma numarasını oku
                    jersey_number = read_jersey_number(jersey_region)
                    if jersey_number != 'N/A':
                        with open(output_csv, "a") as f:
                            f.write(f"{second:.2f},{jersey_number},{bbox}\n")

cap.release()
print(f"Forma numaraları {output_csv} dosyasına kaydedildi.")
# Drive’a kopyala
!cp /content/jersey_numbers.csv /content/drive/MyDrive/
```

```python
%%writefile /content/drive/MyDrive/jersey_number_pipeline/helpers.py
import os
import cv2
import json

def identify_soccer_balls(image_dir, soccer_ball_list):
    print(f"### identify_soccer_balls fonksiyonu başladı ###")
    print(f"image_dir: {image_dir}")

    # image_dir mevcut mu kontrol et
    if not os.path.exists(image_dir):
        print(f"Hata: image_dir ({image_dir}) mevcut değil!")
        return False

    tracklets = os.listdir(image_dir)
    print(f"Tracklets found: {tracklets}")

    if not tracklets:
        print("Hata: image_dir içinde tracklet bulunamadı!")
        ball_json = {'ball_tracks': []}
        with open(soccer_ball_list, 'w') as f:
            json.dump(ball_json, f)
        return False

    ball_tracks = []
    for track in tracklets:
        track_dir = os.path.join(image_dir, track)
        print(f"Track: {track}, Track_dir: {track_dir}")

        # track_dir mevcut mu kontrol et
        if not os.path.isdir(track_dir):
            print(f"Hata: {track_dir} bir dizin değil!")
            continue

        images = os.listdir(track_dir)
        print(f"Images in {track}: {images}")

        if len(images) == 0:
            print(f"Uyarı: {track} klasöründe görüntü bulunamadı!")
            continue

        image_path = os.path.join(track_dir, images[0])
        print(f"Image path: {image_path}")

        # image_path mevcut mu kontrol et
        if not os.path.exists(image_path):
            print(f"Hata: {image_path} dosyası mevcut değil!")
            continue

        image = cv2.imread(image_path)
        if image is None:
            print(f"Hata: {image_path} okunamadı!")
            continue

        try:
            print(f"Image shape: {image.shape}")
            if image.shape[0] < 50 or image.shape[1] < 50:  # Soccer ball images are typically small
                ball_tracks.append(track)
                print(f"{track} bir futbol topu olarak sınıflandırıldı.")
        except Exception as e:
            print(f"Hata: {track} işlenirken bir sorun oluştu: {e}")
            continue

    ball_json = {'ball_tracks': ball_tracks}
    with open(soccer_ball_list, 'w') as f:
        json.dump(ball_json, f)
    print(f"Soccer ball list saved: {soccer_ball_list}")
    print(f"### identify_soccer_balls fonksiyonu tamamlandı ###")
    return True

# Diğer fonksiyonlar
def generate_json(all_files, output_json):
    print(f"### generate_json fonksiyonu başladı ###")
    data = []
    for file_path in all_files:
        data.append({"image_path": file_path})
    with open(output_json, 'w') as f:
        json.dump(data, f)
    print(f"JSON saved: {output_json}")
    print(f"### generate_json fonksiyonu tamamlandı ###")

def evaluate_legibility(gt_path, illegible_path, legible_dict, soccer_ball_list=None):
    print(f"### evaluate_legibility fonksiyonu başladı ###")
    print("Evaluating legibility...")
    print(f"### evaluate_legibility fonksiyonu tamamlandı ###")

def generate_crops(output_json, crops_destination_dir, legible_results):
    print(f"### generate_crops fonksiyonu başladı ###")
    print(f"Generating crops in {crops_destination_dir}...")
    print(f"### generate_crops fonksiyonu tamamlandı ###")

def process_jersey_id_predictions(result_file, useBias=True):
    print(f"### process_jersey_id_predictions fonksiyonu başladı ###")
    with open(result_file, 'r') as f:
        results = json.load(f)
    print(f"### process_jersey_id_predictions fonksiyonu tamamlandı ###")
    return results, None

def evaluate_results(consolidated_dict, gt_dict, full_results=None):
    print(f"### evaluate_results fonksiyonu başladı ###")
    print("Evaluating results...")
    print(f"### evaluate_results fonksiyonu tamamlandı ###")
```

```python
import cv2
import os
import pandas as pd

# Sahne zamanlarını oku
scenes_df = pd.read_csv('/content/drive/MyDrive/match_video-Scenes.csv', header=1)
scenes = scenes_df[['Start Time (seconds)', 'End Time (seconds)']].values

# Video yolu
video_path = '/content/drive/MyDrive/videos/match_video.mp4'
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Kareleri kaydetmek için dizin oluştur
base_output_dir = '/content/drive/MyDrive/jersey_number_pipeline/data/custom_video/images'
os.makedirs(base_output_dir, exist_ok=True)

# Her sahne için kareleri çıkar
for idx, (start_time, end_time) in enumerate(scenes):
    tracklet_dir = f'{base_output_dir}/tracklet_{idx}'
    os.makedirs(tracklet_dir, exist_ok=True)

    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    frame_idx = 0
    for i in range(start_frame, end_frame):
        ret, frame = cap.read()
        if not ret:
            break
        cv2.imwrite(f'{tracklet_dir}/frame_{frame_idx:04d}.jpg', frame)
        frame_idx += 1

cap.release()
print("Kareler çıkarıldı ve kaydedildi.")

# Karelerin kaydedildiğini kontrol et
!ls -R /content/drive/MyDrive/jersey_number_pipeline/data/custom_video/images
```
