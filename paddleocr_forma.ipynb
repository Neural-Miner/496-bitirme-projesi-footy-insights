```python
import cv2
from paddleocr import PaddleOCR
from google.colab.patches import cv2_imshow  # cv2_imshow import et

# OCR modelini başlat
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Dil seçimi ve açı sınıflandırıcıyı aktif et

# Video dosyasını aç
video_path = '/content/2023-2024_11_fraport-tav-antalyaspor_besiktas_cropped (1).mp4'  # Videonun yolu
cap = cv2.VideoCapture(video_path)

# Video işlemesi
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # PaddleOCR ile metin tanıma
    result = ocr.ocr(frame, cls=True)

    # OCR sonucunda sadece rakamları al
    numbers = []
    for line in result[0]:
        text = line[1][0]  # line[1][0] metnin kendisini al
        if text.isdigit():
            numbers.append(text)

    # Eğer numara varsa, ekrana yazdır
    if numbers:
        print("Found numbers:", numbers)

    # Sonuçları ekranda göster (Colab için cv2_imshow kullanıyoruz)
    cv2_imshow(frame)

    # 'q' tuşuna basıldığında çık
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Kaynakları serbest bırak
cap.release()
cv2.destroyAllWindows()
```
